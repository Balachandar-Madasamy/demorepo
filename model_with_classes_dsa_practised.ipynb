```python
from  collections import Counter

def dup(lst):
  s = Counter(lst)
  return [ item for item,count in s.items() if count > 2]
```

```python
print(dup([1,1,2,3,4,5,5,3,3]))
```

```python
def longstr(strn):
  start = max_len = 0
  seen = {}
  for i,char in enumerate(strn):
    if char in seen and seen[char] >= start:
      start = seen[char] + 1
    seen[char] = i
    max_len = max(max_len, i -start+1)

  return max_len
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
class MachineLearning:
  def __init__(self,model,X_train,y_train,X_test,y_test):
    self.model_name = model
    self.X_train = X_train
    self.y_train = y_train
    self.X_test = X_test
    self.y_test = y_test
    self.model = None
    self.scaler = None

  def get_model(self):

    try:
      if self.model_name == 'LR':
        self.scaler = StandardScaler()
        X_train = self.scaler.fit_transform(self.X_train)
        self.model = LogisticRegression(max_iter=1000)
        self.model.fit(X_train,self.y_train)
        return self.model

      elif self.model_name == 'NB':
        self.scaler = StandardScaler()
        X_train = self.scaler.fit_transform(self.X_train)
        self.model = GaussianNB()
        self.model.fit(X_train,self.y_train)
        return self.model

      elif self.model_name == 'DT':
        self.model = DecisionTreeClassifier()
        self.model.fit(self.X_train,self.y_train)
        return self.model

      elif self.model_name == 'RF':
        self.model = RandomForestClassifier()
        self.model.fit(self.X_train,self.y_train)
        return f"{self.model} Model Run Succesfully"

      elif self.model_name == 'XG':
        self.model = xgb.XGBClassifier()
        self.model.fit(self.X_train,self.y_train)
        return self.model

      else:
        return 'Please provide the Valid Model Name to train the model'


    except Exception as e:
      print('Here is the why error is coming',e)

  def __str__(self):
    return f"{self.model_name} Model with parameters Runs Succesfully"

  def __repr__(self):
    return f"{self.model_name} Model with parameters Runs Succesfully"

```

```python
class ModelTrained_Validated(MachineLearning):
  def __init__(self,model,X_train,y_train,X_test,y_test):
    super().__init__(model,X_train,y_train,X_test,y_test)
    self.model_name = model

  def empty_print(self):
    if self.model_name == 'LR':
      act_model_name = 'Logistic Regression'

    elif self.model_name == 'NB':
      act_model_name = 'Naive Bayes'

    elif self.model_name == 'DT':
      act_model_name = 'Decision Tree'

    elif self.model_name == 'RF':
      act_model_name = 'Random Forest'

    elif self.model_name == 'XG':
      act_model_name = 'XGBoost'

    else:
      print("Check the valid model name ")

    print()
    print(f"{act_model_name} Model Results : ")
    print()

  def training_results(self):
    try:

      if self.model_name == 'LR':
        X_train = self.scaler.fit_transform(self.X_train)
        X_test = self.scaler.transform(self.X_test)
        y_pred_test = self.model.predict(X_test)
        train_acc = accuracy_score(self.y_test,y_pred_test)
        self.empty_print()
        print('Training Accuracy for Logistic Regression :',"{:.2f}".format(train_acc * 100),'%')

      elif self.model_name == 'NB':
        X_train = self.scaler.fit_transform(self.X_train)
        X_test = self.scaler.transform(self.X_test)
        y_pred_test = self.model.predict(X_test)
        train_acc = accuracy_score(self.y_test,y_pred_test)
        self.empty_print()
        print('Training Accuracy for Naive Bayes :',"{:.2f}".format(train_acc * 100),'%')

      elif self.model_name == 'RF':
        # X_train = self.scaler.fit_transform(self.X_train)
        # X_test = self.scaler.transform(self.X_test)
        y_pred_test = self.model.predict(self.X_test)
        train_acc = accuracy_score(self.y_test,y_pred_test)
        self.empty_print()
        print('Training Accuracy for Random Forest :',"{:.2f}".format(train_acc * 100),'%')


    except Exception as e:
      print('Metrics has Occured : ',e)
```
